```python
class case_iterator:
    def __init__(self) -> None:

        self.case_lst = []

    def __iter__(self):
        return self
    
    def __next__(self):

        print(self.case_lst)
        if len(self.case_lst) > 0:
            print(self.case_lst[0])

        if np.sum([x.result for x in self.case_lst if x.result]) > 10:
            raise StopIteration

        c = case(np.random.choice(range(10)))
        self.case_lst.append(c)
        return c

```

```python
import multiprocessing
import numpy as np
```

```python
def worker(q_cases, q_results):
    case = q_cases.get()
    case.run()
    q_results.put(case)
```

```python
def generator(q_cases: multiprocessing.Queue, q_results: multiprocessing.Queue, iterator: case_iterator, n_initial):
    
    results = []

    for i in range(n_initial):
        q_cases.put(iterator.__next__())

    

    while np.sum([x.result for x in results]) < 100:
        result = q_results.get()
        results.append(result)

    print(results)

    
```
